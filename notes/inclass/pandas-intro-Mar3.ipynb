```python
names_by_year = pd.DataFrame({'sex':['Female','Male'],
                             2015:['Emma','Noah'],
                             2016:['Emma','Noah'],
                             2017:['Emma','Noah']})
names_by_year
```

```python
# convert from wide to long: use "melt"
names_by_year.melt(id_vars = 'sex')
```

```python
names_melted = names_by_year.melt(id_vars = 'sex',var_name='year',value_name='name')
names_melted
```

```python
# long to wide: use "pivot"
# convert names_melted from long to wide
names_melted.pivot(index='sex',columns='year',values='name')
```
